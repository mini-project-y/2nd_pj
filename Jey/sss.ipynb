```python
import pandas as pd
```

```python
school = pd.read_csv('geocoded_middle.csv', encoding='euc-kr')
```

```python
school
```

```python
school = school.drop_duplicates(subset='address')
```

```python
school.to_csv('uniq_middle.csv', index=False)
```

```python
high = pd.read_csv('geocoded_high.csv', encoding='utf-8')
# euc-kr    utf-8   CP949
```

```python
school[school['no'].str.contains('고등학교')]
```

```python
high
```

```python
high = high.drop_duplicates(subset='address')
high
```

```python
# 데이터프레임 병합
merged_df = pd.merge(
    school, 
    high, 
    on=['no', 'address', 'latitude', 'longitude'],  # 기준 열들
    how='inner'  # 병합 방식 (inner, outer, left, right)
)
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
merged_df.to_csv('uniq_school.csv', index=False)
```

```python
merged_df
```

```python
concated_df = pd.concat([school, high])
concated_df.shape, high.shape, school.shape
```

```python
concated_df
```

```python
concated_df = concated_df.drop_duplicates(subset='address')
concated_df
```

```python
concated_df.to_csv('uniq_all_school.csv', index=False)
```

```python
excel_file = 'uniq_all_school.xlsx'  # 저장할 Excel 파일 이름
concated_df.to_excel(excel_file, index=False, engine='openpyxl')  # engine='openpyxl' 지정

print(f"Excel 파일로 저장되었습니다: {excel_file}")
```
