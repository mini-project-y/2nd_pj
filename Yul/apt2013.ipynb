```python
import pandas as pd

def process_apt_df(df):
    # 시구, 동 컬럼 생성
    df['city_gu'] = df['시군구'].str.split('구 ').str[0] + '구'
    df['dong'] = df['시군구'].str.split('구 ').str[1]

    # 기존주소와 도로명주소 생성
    df['기존주소'] = df['시군구'] + ' ' + df['번지']
    df['도로명주소'] = df['city_gu'] + ' ' + df['도로명']

    # 사용할 컬럼만 추출
    df = df[['NO', '기존주소', '도로명주소']]
    
    return df
```

```python
dataframes = [apt2013_df, apt2014_df, apt2015_df, apt2016_df, apt2017_df, 
              apt2018_df, apt2019_df, apt2020_df, apt2021_df, apt2022_df, 
              apt2023_df, apt2024_df]

processed_dataframes = [process_apt_df(df) for df in dataframes]
# 이후 processed_dataframes는 각 연도별로 처리된 데이터프레임들의 리스트가 됩니다.

apt2013_df, apt2014_df, apt2015_df, apt2016_df, apt2017_df, apt2018_df, apt2019_df, apt2020_df, apt2021_df, apt2022_df, apt2023_df, apt2024_df = processed_dataframes
```
