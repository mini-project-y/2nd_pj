```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/human/2nd_project/경제/경제/서울시 상권분석서비스(소득소비-행정동).csv', encoding='euc-kr')
```

```python
# '기준_년분기_코드'에서 뒷자리 숫자 제거
df['기준_년분기_코드'] = df['기준_년분기_코드'].astype(str).str[:-1]

# 결과 확인
df.head()
```

```python
dfs_by_year = {year: data for year, data in df.groupby('기준_년분기_코드')}

for year, data in dfs_by_year.items():
    print(f"기준_년분기_코드: {year}, 데이터 크기: {data.shape}")
    print(data.head())  # 각 연도의 데이터 미리보기
```

```python
data_of_2019 = dfs_by_year['2019']
new_columns = ['행정동_코드_명','월_평균_소득_금액','지출_총금액','생활용품_지출_총금액','교통_지출_총금액','여가_문화_지출_총금액','음식_지출_총금액'] 
data_of_2019 = data_of_2019[new_columns]
data_of_2019_1 = data_of_2019.groupby('행정동_코드_명').mean()
data_of_2019_1
```
